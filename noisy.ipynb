```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f1=10
f2=100
fs=500
t=np.arange(0,1,1/fs)
x1=np.sin(2*np.pi*f1*t)
x2=np.sin(2*np.pi*f2*t)
x=x1+x2
```

```python
k=np.arange(len(x))
plt.figure()
plt.stem(k,x)
plt.title("Nosiy signal")
plt.grid()
plt.show()
```

```python
fc=20
wc=2*np.pi*fc/fs
N=len(x)
M=(N-1)//2
hd=np.zeros(N)
```

```python
for n in range(N):
    if n==M:
        hd[n]=wc/np.pi
    else:
        hd[n]=np.sin(wc*(n-M))/(np.pi*(n-M))    
```

```python
w=np.zeros(N)
for n in range(N):
    w[n]=0.54-0.46*np.cos((2*np.pi*n)/(N-1))
    
```

```python
h=hd*w
```

```python
plt.figure()
plt.stem(h)
plt.title("Hamming window")
plt.grid()
plt.show()
```

```python
def conv(x,y):
    L=len(x)
    M=len(y)
    N=L+M-1
    y_conv=np.zeros(N)
    for n in range(L):
        for k in range(M):
            if 0<=n-k<M:
                y_conv[n]+=x[n]*y[n-k]
    return y_conv            
    
    
```

```python
value=conv(x,h)
plt.figure() 
plt.plot(value) 
plt.title("Filtered Signal")
plt.grid() 
plt.show()
```
